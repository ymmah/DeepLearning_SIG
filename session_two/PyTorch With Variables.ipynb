```python
import torch
from torch.autograd import Variable

dtype = torch.FloatTensor
batch_size = 64
input_dimension = 1000
hidden_dimension = 100
output_dimension = 10

inputs = Variable(torch.randn(batch_size, input_dimension).type(dtype), requires_grad=False)
targets = Variable(torch.randn(batch_size, output_dimension).type(dtype), requires_grad=False)

input_weights = Variable(torch.randn(input_dimension, hidden_dimension).type(dtype), requires_grad=True)
output_weights = Variable(torch.randn(hidden_dimension, output_dimension).type(dtype), requires_grad=True)

learning_rate = 1e-6
for t in range(500):
  output = inputs.mm(input_weights).clamp(min=0).mm(output_weights)

  loss = (output - targets).pow(2).sum()
  print(t, loss.data[0])

  loss.backward()

  input_weights.data -= learning_rate * input_weights.grad.data
  output_weights.data -= learning_rate * output_weights.grad.data

  input_weights.grad.data.zero_()
  output_weights.grad.data.zero_()
```
