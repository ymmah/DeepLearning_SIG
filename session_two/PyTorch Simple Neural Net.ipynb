```python
# -*- coding: utf-8 -*-

import torch


dtype = torch.FloatTensor
# dtype = torch.cuda.FloatTensor # Uncomment this to run on GPU

# N is batch size; D_in is input dimension;
# H is hidden dimension; D_out is output dimension.
batch_size = 64
input_dimension = 1000
hidden_dimension = 100
output_dimension = 10
epochs = 500
learning_rate = 0.000001

# Create random input and output data
inputs = torch.randn(batch_size, input_dimension).type(dtype)
targets = torch.randn(batch_size, output_dimension).type(dtype)

# Randomly initialize weights
input_weights = torch.randn(input_dimension, hidden_dimension).type(dtype)
output_weights = torch.randn(hidden_dimension, output_dimension).type(dtype)

for t in range(epochs):
    # Forward pass: compute predicted output
    hidden = inputs.mm(input_weights)
    hidden_relu = hidden.clamp(min=0)
    output = hidden_relu.mm(output_weights)

    # Compute and print loss
    loss = (output - targets).pow(2).sum()
    print(t, loss)

    # Backprop to compute gradients of hidden_weights and output_weights with respect to loss
    grad_output = 2.0 * (output - targets)
    grad_output_weights = hidden_relu.t().mm(grad_output)
    grad_hidden_relu = grad_output.mm(output_weights.t())
    grad_h = grad_hidden_relu.clone()
    grad_h[hidden < 0] = 0
    grad_input_weights = inputs.t().mm(grad_h)

    # Update weights using gradient descent
    input_weights -= learning_rate * grad_input_weights
    output_weights -= learning_rate * grad_output_weights
```
