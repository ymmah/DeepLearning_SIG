```python
import numpy as np

# N is batch size; D_in is input dimension;
# H is hidden dimension; D_out is output dimension.
batch_size = 64
input_dimension = 1000
hidden_dimension = 100
output_dimension = 10
epochs = 500
learning_rate = 0.000001

# Create random input and output data
inputs = np.random.randn(batch_size, input_dimension)
targets = np.random.randn(batch_size, output_dimension)

# Randomly initialize weights
hidden_weights = np.random.randn(input_dimension, hidden_dimension)
output_weights = np.random.randn(hidden_dimension, output_dimension)

for t in range(epochs):
    # Forward pass: compute predicted y
    hidden = inputs.dot(hidden_weights)
    hidden_relu = np.maximum(hidden, 0) # make all values positive
    output = hidden_relu.dot(output_weights)

    # Compute and print loss
    error = output - targets
    loss = np.square(error).sum()
    print(t, loss)

    # Backprop to compute gradients of hidden_weights and output_weights with respect to loss
    grad_output = 2.0 * (output - targets)
    grad_output_weights = hidden_relu.T.dot(grad_output)
    grad_hidden_relu = grad_output.dot(output_weights.T)
    grad_hidden = grad_hidden_relu.copy()
    grad_hidden[hidden < 0] = 0
    grad_hidden_weights = inputs.T.dot(grad_hidden)

    # Update weights
    hidden_weights -= learning_rate * grad_hidden_weights
    output_weights -= learning_rate * grad_output_weights

```
